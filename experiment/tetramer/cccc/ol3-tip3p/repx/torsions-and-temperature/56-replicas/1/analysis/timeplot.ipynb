```python
import os, sys
import numpy as np
import pandas as pd
```

### Population

```python
df_list = []
for idx in np.arange(10,110,10):
    filename = os.path.join(f"../analysis_{idx}ns/population.txt")
    #print(filename)
    _df = pd.read_csv(filename, sep='\t')
    _df['TIME'] = idx
    df_list.append(_df)
df = pd.concat(df_list)
```

```python
import matplotlib.pyplot as plt
```

```python
for name in df['NAME'].unique():
    df_sel = df[df['NAME'] == name]
    plt.plot(df_sel['TIME'], df_sel['POPULATION'], marker='o', label=f'{name}')

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xticks(np.arange(10, 110, 10))
plt.xlabel('Time (ns)', fontsize=16)
plt.ylabel('Population', fontsize=16)
#plt.grid()
plt.legend()
#plt.show()
plt.savefig('population_timeplot.png')
```

### J-coupling

```python
df_list = []
for idx in np.arange(10,110,10):
    filename = os.path.join(f"../analysis_{idx}ns/coupling.txt")
    #print(filename)
    _df = pd.read_csv(filename, sep='\t')
    _df['TIME'] = idx
    df_list.append(_df)
df = pd.concat(df_list)
```

```python
#fig, axes = plt.subplots(2, 2, figsize=(12, 12))
for resname in df['RESNAME'].unique():
    df_sel = df[df['RESNAME'] == resname]
    for coupling in df_sel['JCOUPLING'].unique():
        df_sel2 = df_sel[df_sel['JCOUPLING'] == coupling]
        plt.plot(list(df_sel2['TIME']), list(df_sel2['AVERAGE']), marker='o', label=f'{coupling}')
    plt.title(f'{resname}', fontsize=16)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.xticks(np.arange(10, 110, 10))
    plt.xlabel('Time (ns)', fontsize=16)
    plt.ylabel('J-coupling (Hz)', fontsize=16)
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
    plt.tight_layout()
    #plt.show()
    plt.savefig(f'coupling_{resname}_timeplot.png')
    plt.clf()
```
